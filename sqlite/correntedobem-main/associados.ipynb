```python
file = 'associados_vargem_bonita.csv'
df = pd.read_csv(file)
# df = df.iloc[:,1:]

lst = [x for x in df.columns if 'Unnamed' in x]
for x in lst:
    df.drop(x,inplace=True, axis=1)

df.head(3)
```

```python
for cl in df.columns:
    df[cl] = df[cl].apply(lambda x: str(x).upper())
```

```python
lamb_first_name =  lambda x: str(x).split(' ')[0]
lamb_last_name = lambda x: str(x).replace((x.split(' ')[0]),'')
```

```python
def find_numbers(x):
    numbers = re.findall('[0-9]+', str(x))
    numbers = ''.join(numbers)
    return numbers

for cl in [date for date in df.columns if 'data' in date]:
    id = cl.split('_')[-1]
    df['data_de_nascimento_'+id] = df['data_de_nascimento_'+id].apply(find_numbers)
```
